<font color = pink><font size = 5> **NSGA-II**

```python
from pymoo.algorithms.nsga2 import NSGA2
from pymoo.factory import get_problem
from pymoo.optimize import minimize
from pymoo.visualization.scatter import Scatter
from pymoo.util.plotting import plot
import matplotlib.pyplot as plt
```

<font color = pink> <font size = 4> **ZDT1**

```python
zdt1 = get_problem("zdt1")
plt.figure()
x = [i[0] for i in zdt1.pareto_front()]
y = [i[1] for i in zdt1.pareto_front()]
plt.scatter(x, y, label = 'pareto_front')
for i in range(50, 101, 50):
  for j in range(50, 101, 50):
    nsga2 = NSGA2(pop_size = i)
    res = minimize(zdt1, nsga2, ('n_gen', j), seed=18520173, verbose=False)
    lb = "pop_size = " + str(i) + ", n_gen = " + str(j)
    x = [i[0] for i in res.F]
    y = [i[1] for i in res.F]
    plt.scatter(x, y, label = lb)

plt.legend()
```

<font color = pink><font size = 5> **MOEA/D**

```python
from pymoo.algorithms.moead import MOEAD
from pymoo.factory import get_problem, get_visualization, get_reference_directions
from pymoo.optimize import minimize
from pymoo.visualization.scatter import Scatter
from pymoo.util.plotting import plot
import matplotlib.pyplot as plt
```

<font color = pink> <font size = 4> **ZDT1**

```python
zdt1 = get_problem("zdt1")

plt.figure()
x = [i[0] for i in zdt1.pareto_front()]
y = [i[1] for i in zdt1.pareto_front()]
plt.scatter(x, y, label = 'pareto_front')
for j in range(50, 101, 10):
  moead = MOEAD(get_reference_directions("energy", 2, 100), n_neighbors=15, decomposition="tchebi", 
                  prob_neighbor_mating=0.7, seed=18520173)
  res = minimize(zdt1, moead, termination=('n_gen', j))
  lb = "n_gen = " + str(j)
  x = [i[0] for i in res.F]
  y = [i[1] for i in res.F]
  plt.scatter(x, y, label = lb)

plt.legend()
```
